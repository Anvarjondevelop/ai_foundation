```python
import matplotlib.pyplot as plt
import numpy as np
```

# basic Line Plot

```python
x = np.array([1,2,3,4])
y = np.array([10,20,25,30])

plt.plot(x,y)
plt.title("Basic Line Pipe ")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

# Line Styles and Colors

```python
x = np.array([1, 2, 3, 4])
y = np.array([10, 20, 25, 35])

plt.plot(y, linestyle='dotted', color='green', marker='o')
plt.title("Basic Line Example")
plt.show()
```

# Multiplile Line in One graph with Numpy and Matplotlib

```python
x = np.array([1, 2, 3, 4])
y1 = np.array([10, 20, 25, 35])
y2 = np.array([8, 10, 15, 65])

plt.plot(x, y1 ,label='Line 1', color= 'red', marker='s')
plt.plot(x, y2, label = "Line 2" , color='blue', linestyle='--', marker='o')
plt.title("Multiple Line Plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.show()
```

Bar chart with Numpy and Matplotlib

```python
x = np.array(['Money', "Salary", "Annual Salary", "Income"])
y = np.array([2000000,3000000, 9000000, 8000000])

plt.bar(x,y , color ="orange")
plt.title("Basic Bar Chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

# Scatter Plot with Numpy and Matplotlib

```python
x = np.array([1, 2, 3, 4])
y = np.array([10, 40, 25, 35])

plt.scatter(x, y, color='red' , s=50)
plt.title("Scatter Plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

# Histogram  with Numpy and matplotlib

```python
data = np.array([3,4,5,6,7,8,8,8,9])

plt.hist(data, color='green',)
plt.title("Histogram")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.show()
```

# Subplots (Multiple Graphs in One Window)

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4])
y1 = np.array([2, 4, 6, 8])
y2 = np.array([3, 6, 9, 12])

plt.figure(figsize=(8, 4))  # width = 8, height = 4
# 1  => nechta qator? 1 ta qator
# 2  => nechta ustun? 2 ta ustun
# 1  => qaysi joyga chizamiz? 1-joyga
plt.subplot(1, 2, 1)
plt.plot(x, y1, color='red')
plt.title('Plot 1')

plt.subplot(1, 2, 2)
plt.bar(x, y2, color='blue')
plt.title('Plot 2')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4])
y = np.array([4, 8, 6, 10])

plt.plot(x, y, color='green')
plt.title('Save Figure Example')
plt.savefig('my_plot.png')  # Saves in current folder
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4])
y1 = np.array([10, 15, 20, 25])
y2 = np.array([8, 12, 18, 22])

plt.plot(x, y1, label='Dataset A', color='red', marker='o')
plt.plot(x, y2, label='Dataset B', color='blue', marker='s', linestyle='--')

plt.title("Matplotlib with NumPy Arrays")
plt.xlabel("X Values")
plt.ylabel("Y Values")

plt.legend()
plt.grid(True)
plt.show()
```
