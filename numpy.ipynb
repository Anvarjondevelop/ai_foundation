```python
import numpy as np

arr = np.array([1,2,3,4])
print('numpy_arr', arr)
```

```python
arr2 = np.array([[1,2,3,4],[6,7,8,9]])
print('numpy_arr2', arr2)

arr.shape
```

```python
#dtype
a = np.array([1,2,3])
print(a.dtype)
```

```python
a = np.array([1,2,3,4,5,6])
b = a.reshape(2,3)
print('array_reshaped:',b)
```

```python
#astype
a = np.array([1,2,3])
b = a.astype(float)
print(b)
```

```python
# np.zeros(), np.ones() => dummy data
z = np.zeros((2,3))
o = np.ones((2,3))

print(z)
print(o)
```

```python
#ndim
a = np.array([[1,3], [2,4]])
print(a.ndim)
```

```python
#arange
a = np.arange(1,10)
print(a)
```

```python
#sum #mean #min #max
a = np.array([1,2,3,4])

print(a.sum())
print(a.mean())
print(a.min())
print(a.max())
```

```python
#flatten
a = np.array([[1,2,3,4],[5,6,7,8]])
print(a.flatten())
```

```python
# Transpose
a = np.array([[1,2,3,4],[5,6,7,8]])
print(a.T)
```

```python
#dot /@
a = np.array([[1,2],[3,4]])
b = np.array([[4,5],[6,7]])
print('dot/multiplication:', a.dot(b))
```

```python
#random.rand
a = np.random.rand(2,3) # => included 0.0 exluded 1.0
print(a)
```
