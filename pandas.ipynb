```python
import pandas as pd
```

```python
print(pd.__version__)
```

```python
# Dictionary
mydataset = {
  'cars': ['BMW', 'Volvo', 'Ford', 'Tesla', 'Ferrari', 'Pagani', 'BMW'],
  'counts': [3, 7, 2, 5, 8, 6, 29]
}
print(mydataset, "\n")

df = pd.DataFrame(mydataset)
print(df, "\n")
```

```python
df.head(4) # default 5 rows  = > boshidan 5ta qator
```

```python
df.tail(2) # default 5 rows = > oxiridan 5ta qator
```

```python
# State Property: shape, column, dtypes
a = df.shape
print("=====\n", a)

b = df.columns
print("=====\n", b)

c = df.dtypes # har bir column ning type
print("=====\n", c)

df["cars"] = df["cars"].astype("string")
d = df.dtypes
print("=====\n", d)
```

```python
df.describe()
# df.describe(include="all")
```

```python
# iloc (integer location) method

print("=====\n", df.iloc[0])

print("=====\n", df.iloc[1])

print("=====\n", df.iloc[1,0]) # [row location, column location]

print("\n")
df.iloc[1:4]
```

```python

# loc method
df.loc[1:4, "cars"]
```

```python
# mean
a = df.mean(numeric_only=True)
b = df["counts"].mean()

print(a, "\n")
print(b, "\n")

print("======\n")
print(df.min())

print("======\n")
print(df.max())

print("======\n")
print(df.median(numeric_only=True))

print("======\n")
print(df.count())
```
